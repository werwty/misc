```python
# Load the data for all accredited universities from https://ope.ed.gov/dapip/#/home
# and choose two per region

import collections
import csv
import random
import re
```

```python
states = {}


with open('../data/PublicUniversity.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        state = row['State']
        #m = re.search(' ([A-Z]{2})( [0-9]{5}|$)', row['Address'])
        #if m is None:
        #    print(row['Address'])
        #    continue
    
        #state = m.group(1)
        if state not in states:
            states[state] = []
        if row['Campus housing'].lower() == "yes":
            states[state].append(row['Name'])
```

```python
choosen_univ = {}
for state, universities in states.items():
    if len(universities)>=2:
        choosen_univ[state]= random.sample(universities, 2)
    else:
        choosen_univ[state]= universities

    
```

```python
ordeded_univ = collections.OrderedDict(sorted(choosen_univ.items()))
ordeded_univ
```
